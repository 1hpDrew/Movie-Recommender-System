```python
movies_cleaned_df.head(1)['overview']
```

tfidf vectrizer convert text into a vector matrix 


```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3,max_features=None,
                      strip_accents='unicode',analyzer='word',token_pattern=r'\w{1,}',# removing unnecessary symbols like [!,,,.,"",!,.....]
                      ngram_range=(1,3), # take a range of between 1 to 3 kind of words
                      stop_words='english') # this will remove all the unnecessary english words like [the, and, or , a , that , he , she , of .....]

#filling nan with empty string
movies_cleaned_df['overview']=movies_cleaned_df['overview'].fillna('')
```

```python
tfv_matrix=tfv.fit_transform(movies_cleaned_df['overview'])
# this will convert into a sparse matrix
```

```python
#importing sigmoid from sklearn
from sklearn.metrics.pairwise import sigmoid_kernel
#will convert alll values in between 0 to 1

sig=sigmoid_kernel(tfv_matrix,tfv_matrix)
#NOTE :-  yov have to give the same metrics then only combination of this metrics can be done , how summery 1 is related to summery 1 , how summery 1 related o summery 2
```

```python
indices=pd.Series(movies_cleaned_df.index,index=movies_cleaned_df['original_title']).drop_duplicates()
```

```python
indices
```

```python
indices['My Date with Drew']
```

```python
def give_rec(title,sig=sig):
  #get the index corresponding to the original_title
  idx=indices[title]

  #get pairwise similarity scores
  sig_scores = list(enumerate(sig[idx]))

  #sort the movies
  sig_score=sorted(sig_scores,key=lambda x:x[1],reverse =True)

  #score of the 10 most similiar movies
  sig_score=sig_score[1:11]

  #movie indices
  movie_indices=[i[0] for i in sig_score]

  #top 10 most similiar movies
  return movies_cleaned_df['original_title'].iloc[movie_indices]
```

```python
give_rec('The Dark Knight Rises')
```
